# Collazionare con CollateX

### Collazione di un paragrafo del *Lancelot en prose* (mss BnF fr. 16998 e BnF fr. 354) utilizzando [CollateX](http://collatex.net/).
Di seguito si trovano
- gli script che indicano i testi di input, il processo da eseguire e il formato di output
- il risultato della collazione, in formato tabella HTML e in formato grafo
- il risultato della collazione, comparando parola per parola (segmentation=False)

```python
from collatex import *
collation = Collation()
witness_16998 = open( "paragrafo-1-16998.txt", encoding='utf-8' ).read()
witness_354 = open( "paragrafo-1-354.txt", encoding='utf-8' ).read()
witness_micha = open( "paragrafo-1-micha.txt", encoding='utf-8' ).read()
collation.add_plain_witness( "BnF 16998", witness_16998 )
collation.add_plain_witness( "BnF 354", witness_354 )
collation.add_plain_witness( "Ed Micha", witness_micha )
alignment_table = collate(collation, layout='vertical', output='html2')
```
