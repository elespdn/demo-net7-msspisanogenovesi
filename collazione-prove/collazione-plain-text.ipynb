Questa è un esempio di collazione di un paragrafo del *Lancelot en prose* nei mss BnF fr. 16998 e BnF fr. 354, utilizzando [CollateX](http://collatex.net/).

```python
from collatex import *
collation = Collation()
witness_16998 = open( "paragrafo-1-16998.txt", encoding='utf-8' ).read()
witness_354 = open( "paragrafo-1-354.txt", encoding='utf-8' ).read()
witness_micha = open( "paragrafo-1-micha.txt", encoding='utf-8' ).read()
collation.add_plain_witness( "BnF 16998", witness_16998 )
collation.add_plain_witness( "BnF 354", witness_354 )
collation.add_plain_witness( "Ed Micha", witness_micha )
alignment_table = collate(collation, layout='vertical', output='html2')
```
